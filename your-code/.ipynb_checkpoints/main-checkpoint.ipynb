```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=8)
km_model=kmeans.fit(X)
```

```python
from sklearn.metrics import silhouette_score

labels = km_model.labels_
score = silhouette_score(X, labels)
score
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans_5=KMeans(n_clusters=5)
km_model_5=kmeans_5.fit(X)

labels_5 = km_model_5.labels_

score_5 = silhouette_score(X, labels_5)
score_5
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans_3=KMeans(n_clusters=3)
km_model_3=kmeans_3.fit(X)

labels_3 = km_model_3.labels_

score_3 = silhouette_score(X, labels_3)
score_3
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, metric='calinski_harabasz', k=(3,12))
visualizer.fit(X)
visualizer.poof();
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from sklearn.cluster import AgglomerativeClustering

agg_model = AgglomerativeClustering(n_clusters=8)
agg_model.fit(X)

labels = agg_model.labels_

score = silhouette_score(X, labels)
score
```

```python
agg_model_5 = AgglomerativeClustering(n_clusters=5)
agg_model_5.fit(X)

labels_5 = agg_model_5.labels_

score_5 = silhouette_score(X, labels_5)
score_5
```

```python
agg_model_3 = AgglomerativeClustering(n_clusters=3)
agg_model_3.fit(X)

labels_3 = agg_model_3.labels_

score_3 = silhouette_score(X, labels_3)
score_3
```

```python
agg_model = AgglomerativeClustering()
visualizer = KElbowVisualizer(agg_model, k=(3,12))
visualizer.fit(X)
visualizer.poof();
```

```python
km_model = KMeans()
visualizer = KElbowVisualizer(km_model, metric='calinski_harabasz', k=(3,12), locate_elbow=False)
visualizer.fit(X)
visualizer.poof();
```

```python
km_model = KMeans()
visualizer = KElbowVisualizer(km_model, metric='silhouette', k=(3,12))
visualizer.fit(X)
visualizer.poof();
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
X.head(5)
```

```python
kmeans_model = KMeans(n_clusters=3)
kmeans_model.fit(X)
y_pred_km = kmeans_model.predict(X)


plt.figure(figsize=(12,7))
plt.scatter(np.array(X)[:,0], np.array(X)[:,1], c=y_pred_km)

centers = kmeans_model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=50, alpha=0.75)

plt.show();
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components=2)
pca_model.fit(X)
pca_X = pca_model.transform(X)

kmeans_model = KMeans(n_clusters=2)
kmeans_model.fit(pca_X)
y_pred_pca = kmeans_model.predict(pca_X)

plt.figure(figsize=(12,7))
plt.scatter(pca_X[:,0], pca_X[:,1], c=y_pred_pca)

centers = kmeans_model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=50, alpha=0.75)

plt.show();
```
